```python
from sympy import *
import numpy as np
```

```python
Ex, Ey, Ez = map(lambda x:x('z'), symbols('E_x, E_y, E_z', cls=Function))
Hx, Hy, Hz = map(lambda x:x('z'), symbols('H_x, H_y, H_z', cls=Function))
E = Matrix([Ex, Ey, Ez])
H = Matrix([Hx, Hy, Hz])
kx, ky, kz = symbols('k_x, k_y, k_z')
ur = Matrix(np.reshape(symbols('\mu_r1:4(1:4)'), [3, 3]))
er = Matrix(np.reshape(symbols('\epsilon_r1:4(1:4)'), [3, 3]))
```

```python
eq1 = Eq(
    Matrix([
        [I*ky*Ez - Derivative(Ey)],
        [Derivative(Ex)-I*kx*Ez],
        [I*(kx*Ey-ky*Ez)],
    ]),
    ur*H
)

eq2 = Eq(
    Matrix([
        [I*ky*Hz - Derivative(Hy)],
        [Derivative(Hx)-I*kx*Hz],
        [I*(kx*Hy-ky*Hz)],
    ]),
    er*E
)
```

```python
Ex, Ey, Ez = symbols('E_x, E_y, E_z', commutative=False)
Hx, Hy, Hz = symbols('H_x, H_y, H_z', commutative=False)
kx, ky, kz = symbols('k_x, k_y, k_z')
E = Matrix([Ex, Ey, Ez])
H = Matrix([Hx, Hy, Hz])
ur = Matrix(np.reshape(symbols('\mu_r1:4(1:4)', commutative=False), [3, 3]))
er = Matrix(np.reshape(symbols('\epsilon_r1:4(1:4)', commutative=False), [3, 3]))
Hz = ur[2,2]**-1*(I*Matrix([[-ky, kx]])*E[:2, :] - ur[2, :2] * H[:2, :])
Ez = er[2,2]**-1*(I*Matrix([[-ky, kx]])*H[:2, :] - er[2, :2] * E[:2, :])
dEdZ = Matrix([[0,1],[-1,0]])*(
    -I*Matrix([ky, -kx])*Ez+ur[:2, :2]*H[:2, :] + ur[:2, 2]*Hz
)
dHdZ = Matrix([[0,1],[-1,0]])*(
    -I*Matrix([ky, -kx])*Hz+er[:2, :2]*E[:2, :] + er[:2, 2]*Ez
)
M = linear_eq_to_matrix([dEdZ[0], dEdZ[1], dHdZ[0], dHdZ[1]], [Ex, Ey, Hx, Hy])[0]
M = simplify(M)
M
```

```python
Ex, Ey, Ez = symbols('E_x, E_y, E_z', commutative=False)
Hx, Hy, Hz = symbols('H_x, H_y, H_z', commutative=False)
kx, ky, kz = symbols('k_x, k_y, k_z', commutative=False)
E = Matrix([Ex, Ey, Ez])
H = Matrix([Hx, Hy, Hz])
ur = Matrix(np.reshape(symbols('\mu_r1:4(1:4)', commutative=False), [3, 3]))
er = Matrix(np.reshape(symbols('\epsilon_r1:4(1:4)', commutative=False), [3, 3]))
Hz = ur[2,2]**-1*(I*Matrix([[-ky, kx]])*E[:2, :] - ur[2, :2] * H[:2, :])
Ez = er[2,2]**-1*(I*Matrix([[-ky, kx]])*H[:2, :] - er[2, :2] * E[:2, :])
dEdZ = Matrix([[0,1],[-1,0]])*(
    -I*Matrix([ky, -kx])*Ez+ur[:2, :2]*H[:2, :] + ur[:2, 2]*Hz
)
dHdZ = Matrix([[0,1],[-1,0]])*(
    -I*Matrix([ky, -kx])*Hz+er[:2, :2]*E[:2, :] + er[:2, 2]*Ez
)
M = linear_eq_to_matrix([dEdZ[0], dEdZ[1], dHdZ[0], dHdZ[1]], [Ex, Ey, Hx, Hy])[0]
M = simplify(expand(M))
M
```

```python
simplify(expand(M*M))
```

```python
ur_single, er_single = symbols('\mu_r \epsilon_r', commutative=False)
for u in ur:
    M = M.subs(u, ur_single if u.name[-2]==u.name[-1] else '0')
for e in er:
    M = M.subs(e, er_single if e.name[-2]==e.name[-1] else '0', commutative=False)
M
```
